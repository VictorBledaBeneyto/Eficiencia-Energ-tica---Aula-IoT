```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

data_path = 'datos_sensores_temperatura'
archivos = [f for f in os.listdir(data_path) if f.endswith('.csv')]

df_master_temp = pd.DataFrame()

for archivo in archivos:
    df_temp = pd.read_csv(os.path.join(data_path, archivo), na_values=['--'])
    
    df_temp['timestamp'] = pd.to_datetime(df_temp['date'].astype(str) + ' ' + df_temp['time'].astype(str), format='mixed')
    
    df_temp = df_temp[['timestamp', 'temperature ℃']].rename(columns={'temperature ℃': archivo.replace('.csv', '')})
    
    if df_master_temp.empty:
        df_master_temp = df_temp
    else:
        df_master_temp = pd.merge(df_master_temp, df_temp, on='timestamp', how='inner')

matriz_corr_temp = df_master_temp.drop(columns=['timestamp']).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(matriz_corr_temp, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación entre Temperatura')
plt.show()
```

# Temperatura del Aula (Media)

```python
data_path = "../data/silver/sensores_filtrados_limpios.csv"

df = pd.read_csv(data_path)

df = df[df['sensor'] == 'temp_interior']

df['time'] = pd.to_datetime(df['time'])

df = df.sort_values('time')

plt.figure(figsize=(10, 8))

plt.plot(df['time'], df['valor'], 
         label='Temperatura Interior', 
         color='coral', 
         linewidth=1.5)

plt.title('Temperatura Interior del Aula', fontsize=14)
plt.xlabel('Fecha y Hora', fontsize=12)
plt.ylabel('Temperatura (°C)', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

plt.show()
```
